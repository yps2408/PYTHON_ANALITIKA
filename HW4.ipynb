Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python

sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
```

Видна довольно четкая линейная зависимость, хотя мы видим, что диапазон для выбора коэф-та угла наклона довольно широкий

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
```

С ростом оценки стоимость недвижимости растет, что вполне ожидаемо.

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')

plt.xlabel('Кол-во комнат')
plt.ylabel('Стоимость ');
```

Сложно интерпретировать график, видим, что дома с большим количеством комнат (например, 7) могут стоить столько же, сколько дома с 3 комнатами. Поэтому сложно сделать вывод.

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
```

Здесь мы видим более выраженную зависимость, линейную.

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

```python
plt.figure(figsize=(6, 4))

data4 = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'condition', 'grade', 'floors', 'waterfront', 'view']]
sns.heatmap(data4.corr(), annot=True, linewidth=.5, cmap=sns.cubehelix_palette(as_cmap=True))
```
